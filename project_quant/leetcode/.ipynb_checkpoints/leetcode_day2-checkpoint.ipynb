3. 无重复字符的最长子串
给定一个字符串 s ，请你找出其中不含有重复字符的 最长 子串 的长度。

 

示例 1:

输入: s = "abcabcbb"
输出: 3 
解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。注意 "bca" 和 "cab" 也是正确答案。
示例 2:

输入: s = "bbbbb"
输出: 1
解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。
示例 3:

输入: s = "pwwkew"
输出: 3
解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。
     请注意，你的答案必须是 子串 的长度，"pwke" 是一个子序列，不是子串

```python
class Solution:
    def lengthOfLongestSubstring(self, s:str)->int:
        occ=set() #初始化已见参数集合
        n=len(s)
        rk, ans = -1, 0 #初始化右指针
        for i in range(n):
            if i!=0:
                occ.remove(s[i]-1) #当左指针从第一个元素向右移动时，移除指针左边的元素
            while rk+1<n and s[rk+1] not in occ: #判断元素没有超出范围且没有出现在表中
                occ.add(s[rk+1])
                rk+=1 #指针右移
        ans = max(ans, rk-i+1)
```

给定两个字符串 s 和 p，找到 s 中所有 p 的 异位词 的子串，返回这些子串的起始索引。不考虑答案输出的顺序。
示例 1:

输入: s = "cbaebabacd", p = "abc"
输出: [0,6]
解释:
起始索引等于 0 的子串是 "cba", 它是 "abc" 的异位词。
起始索引等于 6 的子串是 "bac", 它是 "abc" 的异位词。
 示例 2:

输入: s = "abab", p = "ab"
输出: [0,1,2]
解释:
起始索引等于 0 的子串是 "ab", 它是 "ab" 的异位词。
起始索引等于 1 的子串是 "ba", 它是 "ab" 的异位词。
起始索引等于 2 的子串是 "ab", 它是 "ab" 的异位词。

```python
class Solution:
    def findAnagrams(self, s:str, p:str) ->List[int]:
        s_len, p_len = len(s), len(p)
        p_counter=Counter(p) #获得p的计数字典
        window_counter=Counter(s[:p_len]) #获得window的计数字典
        res = []
        if p_counter==window_counter:
            res.append(0)
        for i in range(p_len, s_len):
            window_counter[s[i]]+=1 #window计数字典加入新增元素的个数
            left_char=s[i-p_len] 
            window_counter[s[left_char]]-=1 #指针右移，指针左侧元素移出子字典
            if window_counter[s[left_char]]==0: #当元素计数为0时，移除元素
                del window_counter[s[left_char]] 
            if window_counter == p_counter: #动态判断是否窗口对齐，对齐后获得左侧元素加入res中
                res.append(i-p_len+1) 
        return res
```
