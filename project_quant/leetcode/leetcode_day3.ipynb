给定两个字符串 s 和 t，长度分别是 m 和 n，返回 s 中的 最短窗口 子串，使得该子串包含 t 中的每一个字符（包括重复字符）。如果没有这样的子串，返回空字符串 ""。

测试用例保证答案唯一。

 

示例 1：

输入：s = "ADOBECODEBANC", t = "ABC"
输出："BANC"
解释：最小覆盖子串 "BANC" 包含来自字符串 t 的 'A'、'B' 和 'C'。
示例 2：

输入：s = "a", t = "a"
输出："a"
解释：整个字符串 s 是最小覆盖子串。
示例 3:

输入: s = "a", t = "aa"
输出: ""
解释: t 中两个字符 'a' 均应包含在 s 的子串中，
因此没有符合条件的子字符串，返回空字符串。
 

提示：

m == s.length
n == t.length
1 <= m, n <= 105
s 和 t 由英文字母组成
 

进阶：你能设计一个在 O(m + n) 时间内解决此问题的算法吗？

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        counter_t=Counter(t)
        required=len(counter_t) #记录当前窗口满足t要求的字符种类数
        left = 0
        formed = 0 #当前窗口内有多少字符已经达到了target_counts的要求
        ans = (float('inf'), None, None) #记录最小的窗口长度以及左右边界
        window_count={}
        for right, char in enumerate(s):
            #1.扩展窗口：如果当前字符是t里的字符，更新窗口
            if char in counter_t:
                window_count[char]=window_count.get(char, 0)+1
                # 如果这个字符的数量刚好达到了t中的要求，formed加1
                if window_count[char] == counter_t[char]:
                    formed+=1
            #2. 收缩窗口：当窗口已经包含了所有t中的字符时，尝试缩小左边界
            while left <= right and formed == required:
                #更新全局最小窗口
                if right-left-1<ans[0]:
                    ans = (right-left-1, left, right)
                #准备丢弃左指针指向的字符
                removed_char=s[left]
                if removed_char in counter_t:
                    if window_count[removed_char]==counter_t[removed_char]:
                        formed-=1
                    window_count[removed_char]-=1
                left+=1
        return "" if ans[0]==float('inf') else s[ans[1]:ans[2]+1]
```
