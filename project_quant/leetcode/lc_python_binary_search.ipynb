### 二分查找"闭区间"统一写法约定

本笔记统一使用**闭区间 [lk, rk]**：
- 初始化：`lk, rk = 0, n-1`
- 循环条件：`while lk <= rk`（旋转最小值例外，见 LC 153）
- 收缩右边界：`rk = mid - 1`（mid 已检查，排除）
- 收缩左边界：`lk = mid + 1`

> ⚠️ 两种写法（闭区间 vs 左闭右开）**不能混用**，否则死循环或漏解。

## 1. LC 704 — 二分查找（标准三分支）

| 项目 | 内容 |
|------|------|
| **题型** | 标准二分查找（三分支：<、>、==）|
| **时间复杂度** | O(log n) |
| **空间复杂度** | O(1) |
| **数据范围** | 1 ≤ n ≤ 10⁴，−9999 ≤ nums[i], target ≤ 9999，升序不重复 |

### ⚡ 触发条件
1. 数组**已排序**（存在单调性）
2. 要求 **O(log n)**（强制提示二分）
3. 查找某值**是否存在**（存在返回下标，不存在返回 -1）

### 🧠 算法
```
lk, rk = 0, n-1
while lk <= rk:          # 闭区间，lk==rk 仍需检查
    mid = (lk+rk)//2
    if target > nums[mid]:   lk = mid+1   # 偏大，收缩左边界
    elif target < nums[mid]: rk = mid-1   # 偏小，收缩右边界
    else:                    return mid   # 命中
return -1                  # 区间为空，未找到
```

**mid 计算**：`(lk+rk)//2` 在 Python 中无溢出风险；Java/C++ 建议 `lk + (rk-lk)//2`。

````python
# 模板：标准二分查找（闭区间）
def binary_search_template(nums: List[int], target: int) -> int:
    lk, rk = 0, len(nums)-1
    while lk <= rk:
        mid = (lk+rk)//2
        if target > nums[mid]:   lk = mid+1
        elif target < nums[mid]: rk = mid-1
        else:                    return mid
    return -1
print("✅ 标准二分模板")
````

````python
class Solution_704:
    def search(self, nums: List[int], target: int) -> int:
        lk, rk = 0, len(nums)-1
        while lk <= rk:
            mid = (lk+rk)//2
            if target > nums[mid]:   lk = mid+1
            if target < nums[mid]:   rk = mid-1
            if target == nums[mid]:  return mid
        return -1

sol704 = Solution_704()
cases = [([-1,0,3,5,9,12],9,4),([-1,0,3,5,9,12],2,-1),([-1,0,3,5,9,12],-1,0),([-1,0,3,5,9,12],12,5),([5],5,0),([5],3,-1)]
for nums,t,e in cases:
    r = sol704.search(nums,t)
    print(f"{'✅' if r==e else '❌'}  target={t} → {r} (期望 {e})")
````

## 2. LC 35 — 搜索插入位置（Lower Bound）

| 项目 | 内容 |
|------|------|
| **题型** | 二分查找 — Lower Bound（第一个 ≥ target 的下标）|
| **时间复杂度** | O(log n) |
| **空间复杂度** | O(1) |
| **数据范围** | 1 ≤ n ≤ 10⁴，−10⁴ ≤ nums[i], target ≤ 10⁴，升序不重复 |

### ⚡ 触发条件
1. 有序数组 + O(log n)（同 LC 704）
2. **不只是"存在与否"**，还需要确定边界位置（即使 target 不存在）
3. 关键词："第一个 ≥/>","最后一个 ≤/<","插入位置"

### 🧠 算法（与 LC 704 主循环**完全相同**，仅末尾不同）
```
lk, rk = 0, n-1
while lk <= rk:
    mid = (lk+rk)//2
    if target > nums[mid]:    lk = mid+1
    elif target == nums[mid]: return mid
    else:                      rk = mid-1
return lk    # ← 唯一区别：return lk（插入位置），而非 -1
```

**为何 `return lk` 是正确插入位置？**  
循环结束时 lk = rk+1，`nums[0..rk]` 全部 < target，`nums[lk..n-1]` 全部 ≥ target。  
所以 lk 就是"第一个 ≥ target 的下标" = 应插入的位置。

| 需求 | 返回值 |
|------|--------|
| 第一个 ≥ target（本题）| `lk` |
| 第一个 > target | 把 `==` 分支归入"偏小"，同样 `return lk` |
| 最后一个 < target | `lk - 1` |

````python
class Solution_35:
    def searchInsert(self, nums: List[int], target: int) -> int:
        lk, rk = 0, len(nums)-1
        while lk <= rk:
            mid = (lk+rk)//2
            if target > nums[mid]:    lk = mid+1
            elif target == nums[mid]: return mid
            else:                     rk = mid-1
        return lk   # 插入位置

sol35 = Solution_35()
cases = [([1,3,5,6],5,2),([1,3,5,6],2,1),([1,3,5,6],7,4),([1,3,5,6],0,0),([],5,0),([1],1,0),([1],2,1)]
for nums,t,e in cases:
    r = sol35.searchInsert(nums,t)
    print(f"{'✅' if r==e else '❌'}  {nums}, target={t} → {r} (期望 {e})")
````

## 3. LC 153 — 寻找旋转排序数组中的最小值

| 项目 | 内容 |
|------|------|
| **题型** | 旋转数组二分（找最小值/旋转点）|
| **时间复杂度** | O(log n) |
| **空间复杂度** | O(1) |
| **数据范围** | 1 ≤ n ≤ 5000，−5000 ≤ nums[i] ≤ 5000，元素互不相同 |

### ⚡ 触发条件
1. **旋转排序数组**（原升序，从某点截断重接）
2. 要求 O(log n)（提示二分）
3. 查找**最小值**（即找旋转点/断层位置）

### 🧠 算法（与标准二分的关键不同）

**与 LC 704/35 的三处核心差异**：

| | LC 704/35（标准）| LC 153（旋转最小值）|
|--|--|--|
| 比较对象 | `nums[mid]` vs `target` | `nums[mid]` vs `nums[rk]`（没有target！）|
| 循环条件 | `while lk <= rk` | **`while lk < rk`**（收敛到单元素退出）|
| 收缩右边界 | `rk = mid-1` | **`rk = mid`**（mid 本身可能是最小值！）|

```
lk, rk = 0, n-1
while lk < rk:                 # 注意：< 不是 <=
    mid = (lk+rk)//2
    if nums[mid] > nums[rk]:
        lk = mid + 1           # mid 在左半段（较高），最小值在右
    else:                       # nums[mid] <= nums[rk]
        rk = mid               # mid 在右半段（较低），最小值在 mid 或其左
                                # 注意：rk=mid 而非 mid-1！
return nums[lk]                 # 此时 lk == rk

```

**为何比较 `nums[mid]` 与 `nums[rk]`（而非 `nums[lk]`）？**  
旋转后 `nums[rk]` 始终是"右半段结尾"，能稳定区分 mid 在哪一段。  
若 `nums[mid] > nums[rk]`，mid 必在左半（较大区），最小值在右；  
否则 mid 在右半（较小区），最小值在 [lk, mid]。

````python
# 模板：旋转数组找最小值
def find_min_rotated(nums: List[int]) -> int:
    lk, rk = 0, len(nums)-1
    while lk < rk:              # 收敛到单个元素
        mid = (lk+rk)//2
        if nums[mid] > nums[rk]:
            lk = mid+1          # 最小值在右半
        else:
            rk = mid            # 最小值在左半（含 mid）
    return nums[lk]             # lk == rk 时即为最小值
print("✅ 旋转数组找最小值模板")
````

````python
class Solution_153:
    def findMin(self, nums: List[int]) -> int:
        lk, rk = 0, len(nums)-1
        while lk < rk:
            mid = (lk+rk)//2
            if nums[mid] > nums[rk]:
                lk = mid+1
            else:
                rk = mid
        return nums[lk]

sol153 = Solution_153()
cases = [([3,4,5,1,2],1),([4,5,6,7,0,1,2],0),([11,13,15,17],11),([1],1),([2,1],1),([5,1,2,3,4],1)]
for nums,e in cases:
    r = sol153.findMin(nums)
    print(f"{'✅' if r==e else '❌'}  {nums} → {r} (期望 {e})")
````

## 4. LC 33 — 搜索旋转排序数组（找目标值）

| 项目 | 内容 |
|------|------|
| **题型** | 旋转数组二分（找目标值）|
| **时间复杂度** | O(log n) |
| **空间复杂度** | O(1) |
| **数据范围** | 1 ≤ n ≤ 5000，−10⁴ ≤ nums[i], target ≤ 10⁴，元素互不相同 |

### ⚡ 触发条件
1. **旋转排序数组**，需查找某个目标值
2. 要求 O(log n)

### 🧠 算法（核心洞察：每次迭代必有一半是有序的）

```
lk, rk = 0, n-1
while lk <= rk:
    mid = (lk+rk)//2
    if nums[mid] == target: return mid

    if nums[lk] <= nums[mid]:          # 左半 [lk, mid] 有序（注意：<=，覆盖 lk==mid 的情况）
        if nums[lk] <= target < nums[mid]: rk = mid-1   # target 在有序左半内
        else:                              lk = mid+1   # target 在无序右半
    else:                               # 右半 [mid, rk] 有序
        if nums[mid] < target <= nums[rk]: lk = mid+1   # target 在有序右半内
        else:                              rk = mid-1   # target 在无序左半
return -1
```

**关键：`nums[lk] <= nums[mid]`（为何用 <=）**  
当只剩 2 个元素时 `lk + 1 = rk`，mid = lk，`nums[lk] == nums[mid]`，  
此时"左半"只有一个元素 mid，天然有序。用 `<=` 才能正确进入左半分支。

**LC 33 vs LC 153 对比**：

| | LC 153 找最小值 | LC 33 找目标值 |
|--|--|--|
| 比较对象 | `nums[mid]` vs `nums[rk]` | `nums[lk]` vs `nums[mid]`（识别有序半段）|
| 循环条件 | `while lk < rk` | `while lk <= rk` |
| 右边界收缩 | `rk = mid` | `rk = mid-1` |
| 未找到时 | `return nums[lk]` | `return -1` |

````python
class Solution_33:
    def search(self, nums: List[int], target: int) -> int:
        n = len(nums)
        lk, rk = 0, n-1
        while lk <= rk:
            mid = (lk+rk)//2
            if nums[mid] == target: return mid

            if nums[lk] <= nums[mid]:              # 左半有序
                if nums[lk] <= target < nums[mid]:
                    rk = mid-1
                else:
                    lk = mid+1
            else:                                   # 右半有序
                if nums[mid] < target <= nums[rk]:
                    lk = mid+1
                else:
                    rk = mid-1
        return -1

sol33 = Solution_33()
cases = [([4,5,6,7,0,1,2],0,4),([4,5,6,7,0,1,2],3,-1),([1],0,-1),([1],1,0),([5,1],1,1),([5,1],5,0),([3,1,2],1,1)]
for nums,t,e in cases:
    r = sol33.search(nums,t)
    print(f"{'✅' if r==e else '❌'}  {nums}, target={t} → {r} (期望 {e})")
````

## 📋 总结：二分查找全模板

### 四题对照表

| 题号 | 题型 | 时间 | 循环条件 | 比较对象 | 右边界收缩 | 未命中返回 | 易错点 |
|------|------|------|---------|---------|----------|----------|-------|
| 704 | 标准二分 | O(log n) | `lk<=rk` | vs target | `mid-1` | `-1` | <= 与 mid-1 配套 |
| 35  | Lower Bound | O(log n) | `lk<=rk` | vs target | `mid-1` | **`lk`** | 未命中返 lk 非 -1 |
| 153 | 旋转最小值 | O(log n) | **`lk<rk`** | vs `nums[rk]` | **`mid`** | `nums[lk]` | rk=mid非mid-1 |
| 33  | 旋转查找 | O(log n) | `lk<=rk` | vs `nums[lk]`/`nums[rk]` | `mid-1` | `-1` | <= 处理两元素情况 |

### 🔁 所有模板（闭区间）

```python
# 1. 标准二分（LC 704）
lk, rk = 0, n-1
while lk <= rk:
    mid = (lk+rk)//2
    if t > a[mid]:   lk = mid+1
    elif t < a[mid]: rk = mid-1
    else:            return mid
return -1

# 2. Lower Bound（LC 35）— 仅末行不同
lk, rk = 0, n-1
while lk <= rk:
    mid = (lk+rk)//2
    if t > a[mid]:    lk = mid+1
    elif t == a[mid]: return mid
    else:             rk = mid-1
return lk          # ← 插入位置

# 3. 旋转数组找最小值（LC 153）
lk, rk = 0, n-1
while lk < rk:     # ← 注意 <
    mid = (lk+rk)//2
    if a[mid] > a[rk]: lk = mid+1
    else:              rk = mid   # ← mid 非 mid-1
return a[lk]

# 4. 旋转数组找目标值（LC 33）
lk, rk = 0, n-1
while lk <= rk:
    mid = (lk+rk)//2
    if a[mid] == t: return mid
    if a[lk] <= a[mid]:               # 左半有序（<=覆盖两元素情况）
        if a[lk] <= t < a[mid]: rk = mid-1
        else:                    lk = mid+1
    else:                             # 右半有序
        if a[mid] < t <= a[rk]: lk = mid+1
        else:                    rk = mid-1
return -1
```
