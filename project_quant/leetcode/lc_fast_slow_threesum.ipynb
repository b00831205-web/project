## 1. LC 26 — 删除有序数组中的重复项（快慢指针）

| 字段 | 内容 |
|------|------|
| **题号** | 26 |
| **难度** | 🟢 简单 |
| **题型** | 快慢指针 — 读写双指针（Read / Write Two Pointers）|
| **时间复杂度** | O(n) |
| **空间复杂度** | O(1) |
| **数据范围** | 1 ≤ n ≤ 3×10⁴，−100 ≤ nums[i] ≤ 100，数组非递减有序 |

### 题目核心
**原地**删除有序数组的重复元素，使每个元素只出现一次，返回新长度 k。  
前 k 个元素应为排序后的唯一值，其余位置可忽略。

### ⚡ 快慢指针的两种范式（重要区分）

快慢指针在不同场景下有两种完全不同的工作机制：

| 范式 | 代表题 | slow 含义 | fast 含义 | 速度关系 | 目的 |
|------|--------|-----------|-----------|---------|------|
| **读写双指针** | LC 26, LC 80, LC 283 | 写指针（有效区末尾）| 读指针（当前扫描位置）| fast 总是前进；slow **条件性**前进 | 原地筛选 / 覆盖数组 |
| **追及双指针** | LC 141, LC 142, LC 876 | 慢指针（每次 1 步）| 快指针（每次 2 步）| fast **固定**比 slow 快 1 步 / 次 | 检测环 / 找链表中点 |

> 核心区别：  
> - 读写版：速度差来自**条件判断**（发现新元素时 slow 才移动）  
> - 追及版：速度差是**固定的**（fast 永远每次多走 1 步）

---

### LC 26 — 读写双指针

#### ⚡ 触发条件

| 信号 | 说明 |
|------|------|
| 原地修改数组，O(1) 额外空间 | 不能 return 新数组 |
| 需要跳过重复 / 特定值 | slow 维护有效区，fast 扫描全部 |
| 有序数组，重复元素必相邻 | 只需比较 fast 与 slow，无需哈希表 |

#### 🧠 算法

```
slow = 0   （写指针：有效区最后一个元素的位置）

for fast in range(1, n):
    if nums[fast] != nums[slow]:   # 发现新的唯一元素
        slow += 1                   # 有效区向右扩展一位
        nums[slow] = nums[fast]     # 将新元素写入有效区末尾

return slow + 1   （有效元素数量 = 最后有效索引 + 1）
```

**区域划分直觉**：

```
nums:  [0, 0, 1, 1, 1, 2, 2, 3]
        ↑              ↑
       slow           fast（扫描中）

nums[0..slow]       = 已确认的有效唯一区
nums[slow+1..fast)  = 已扫过但跳过的重复区（可覆盖）
nums[fast..]        = 待扫描区
```

### LC 141 — 追及双指针（对照示例）

#### ⚡ 触发条件

| 信号 | 题目 |
|------|------|
| 链表是否存在环 | LC 141（本例）|
| 找链表的中间节点 | LC 876（fast 到末尾时 slow 正好在中点）|
| 找环的入口节点 | LC 142（Floyd 算法进阶）|

#### 🧠 算法（Floyd 判圈）

```
slow, fast = head, head

while fast and fast.next:
    slow = slow.next         # 慢指针：走 1 步
    fast = fast.next.next    # 快指针：走 2 步
    if slow is fast:          # 相遇 → 有环
        return True

return False   （fast 提前到达 None → 无环）
```

**Floyd 证明直觉**：有环时，fast 在环内追赶 slow，每次迭代缩短相对距离 1 步，  
必在有限步内相遇（最多循环一圈）。无环时 fast 先到达 None。

> ⚠️ 用 `slow is fast`（对象同一性），不用 `slow == fast`（值相等），  
> 避免两个不同节点恰好值相同时产生误判。

````python
# ===== 读写双指针通用模板 =====
# 触发：原地修改有序数组 + 跳过重复/特定值 + O(1) 空间
# 复杂度：O(n) time | O(1) space

def read_write_ptr_demo(nums: List[int]) -> int:
    # 以 LC 26（去重，保留 1 次）为例
    slow = 0  # 写指针
    for fast in range(1, len(nums)):
        if nums[fast] != nums[slow]:    # keep_condition：发现新唯一值
            slow += 1
            nums[slow] = nums[fast]
    return slow + 1

# ── 不同题目的 keep_condition 对照 ────────────────────────────
print("keep_condition 对照：")
print("  LC 26  (去重，保留1次) :  nums[fast] != nums[slow]")
print("  LC 80  (去重，保留2次) :  nums[fast] != nums[slow-1]  ← 注意是 slow-1")
print("  LC 283 (移零到末尾)    :  nums[fast] != 0")
print()
print("✅ 读写双指针模板已加载")
````

````python
# ===== LC 26 完整解法 =====
# O(n) time | O(1) space
# slow = 写指针（有效区末尾）；fast = 读指针（扫描位置）

class Solution_26:
    def removeDuplicates(self, nums: List[int]) -> int:
        slow = 0   # 有效区最后一个元素的下标

        for fast in range(1, len(nums)):
            if nums[fast] != nums[slow]:   # 发现新的唯一元素
                slow += 1                   # 扩展有效区
                nums[slow] = nums[fast]     # 写入

        return slow + 1   # 有效区长度（下标从 0 开始，故 +1）

print("✅ LC 26 解法已加载")
````

````python
# ===== LC 141 — 链表判环（追及双指针，对照示例）=====
# O(n) time | O(1) space — Floyd 判圈算法

class ListNode:
    def __init__(self, val=0):
        self.val = val
        self.next = None
    def __repr__(self):
        return f"ListNode({self.val})"

class Solution_141:
    def hasCycle(self, head: ListNode) -> bool:
        slow, fast = head, head

        while fast and fast.next:
            slow = slow.next         # 慢指针：每次走 1 步
            fast = fast.next.next    # 快指针：每次走 2 步
            if slow is fast:          # 同一节点 → 有环
                return True

        return False   # fast 到达末尾（None）→ 无环


# 辅助函数：构造链表（cycle_pos >= 0 时创建环）
def make_linked_list(vals: list, cycle_pos: int = -1) -> ListNode:
    if not vals:
        return None
    nodes = [ListNode(v) for v in vals]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
    if cycle_pos >= 0:
        nodes[-1].next = nodes[cycle_pos]   # 尾节点指向 cycle_pos，形成环
    return nodes[0]

print("✅ LC 141 解法 + 辅助函数已加载")
````

````python
# ===== 测试用例：LC 26 =====

sol26 = Solution_26()

tests26 = [
    ([1, 1, 2],                 2, [1, 2]),
    ([0, 0, 1, 1, 1, 2, 2, 3],  4, [0, 1, 2, 3]),
    ([1],                       1, [1]),
    ([-3, -2, -2, 0, 1],        4, [-3, -2, 0, 1]),
]

all_pass = True
for nums, expected_k, expected_vals in tests26:
    nums_copy = nums.copy()
    k = sol26.removeDuplicates(nums_copy)
    actual = nums_copy[:k]
    ok = k == expected_k and actual == expected_vals
    if not ok:
        all_pass = False
    flag = "✅" if ok else "❌"
    print(f"{flag}  输入={nums}")
    print(f"      k={k}, 前{k}项={actual}  (期望 k={expected_k}, vals={expected_vals})")

print()
print("🎉 全部通过！" if all_pass else "⚠️ 有用例未通过")
````

````python
# ===== 测试用例：LC 141 链表判环 =====

sol141 = Solution_141()

tests141 = [
    # (vals,          cycle_pos,  expected)
    ([3, 1, 2, 0],    1,          True),    # 3→1→2→0→1（环），尾→index 1
    ([1, 2],         -1,          False),   # 1→2→None，无环
    ([1],             0,          True),    # 单节点自环：1→1
    ([],             -1,          False),   # 空链表
    ([1, 2, 3, 4],   -1,          False),   # 线性链表，无环
]

all_pass = True
for vals, cycle_pos, expected in tests141:
    head = make_linked_list(vals, cycle_pos)
    result = sol141.hasCycle(head)
    ok = result == expected
    if not ok:
        all_pass = False
    flag = "✅" if ok else "❌"
    desc = f"尾→index {cycle_pos}" if cycle_pos >= 0 else "无环"
    print(f"{flag}  vals={str(vals):<18}  {desc:<15}  →  {result}  (期望 {expected})")

print()
print("🎉 全部通过！" if all_pass else "⚠️ 有用例未通过")
````

## 3. LC 15 — 三数之和（排序 + 对撞双指针）

| 字段 | 内容 |
|------|------|
| **题号** | 15 |
| **难度** | 🟡 中等 |
| **题型** | 排序 + 对撞双指针 |
| **时间复杂度** | O(n²)：排序 O(n log n) + 双层循环 O(n²) |
| **空间复杂度** | O(1)（不计输出，排序原地）|
| **数据范围** | 3 ≤ n ≤ 3000，−10⁵ ≤ nums[i] ≤ 10⁵ |

### 题目核心
找出数组中所有满足 `a + b + c = 0` 的三元组（i ≠ j ≠ k），  
**答案不含重复三元组**（顺序无关）。

### ⚡ 触发条件

| 信号 | 操作 |
|------|------|
| 找**三个元素**之和 = 目标值 | 排序 + 外层固定一个，内层转两数之和 |
| 答案**不含重复三元组** | 三个维度均需去重 |
| 暴力 O(n³) 超时 | 排序后降至 O(n²) |

---

### 🧠 降维思想：三数 → 二数

**核心**：先排序，外层固定 `nums[i]`，问题转化为在 `nums[i+1..n-1]` 中找：

> `nums[left] + nums[right] = −nums[i]`

这正是 **LC 167（两数之和 II）** 的形式！利用有序性用对撞双指针解决。

```
nums.sort()

for i in range(n):
    # ── 剪枝：已排序，nums[i] > 0 → 三数之和必然 > 0，直接终止
    if nums[i] > 0: break

    # ── 去重①（外层 i）：与上一个 i 相同，跳过
    if i > 0 and nums[i] == nums[i-1]: continue

    left, right = i+1, n-1
    while left < right:
        total = nums[i] + nums[left] + nums[right]

        if total == 0:
            记录 [nums[i], nums[left], nums[right]]
            # ── 去重②（left）：跳过 left 的重复值，再前进一步
            while left < right and nums[left] == nums[left+1]:  left += 1
            # ── 去重③（right）：跳过 right 的重复值，再后退一步
            while left < right and nums[right] == nums[right-1]: right -= 1
            left += 1;  right -= 1

        elif total < 0: left  += 1   # 和偏小，右移左指针
        else:           right -= 1   # 和偏大，左移右指针
```

---

### 🔍 三层去重详解（面试高频考点）

| 去重层 | 代码 | 原因 |
|--------|------|------|
| **外层 i** | `if i > 0 and nums[i] == nums[i-1]: continue` | 相同 nums[i] 已枚举过所有配对，再枚举会产生重复三元组 |
| **left 去重** | `while nums[left] == nums[left+1]: left += 1` | 记录解后，left 下一个若值相同会重复 |
| **right 去重** | `while nums[right] == nums[right-1]: right -= 1` | 同理 |

> ⚠️ **常见错误**：外层去重写成 `nums[i] == nums[i+1]`（向后比较）  
> 应向**前**比较 `i-1`。原因：`nums[i+1]` 还未处理，不能因它与 i 相同就跳过当前 i。  
> **口诀**：问"这个值**之前**处理过吗" → 看 `i-1`，不是 `i+1`。

---

### 与 LC 167 的关系

| | LC 167 | LC 15 内层双指针 |
|--|--|--|
| 数组 | 已给定有序数组 | 排序后的 nums[i+1..n-1] |
| 目标和 | 给定 target | −nums[i]（由外层决定）|
| 返回 | 单个下标对 | 所有满足条件的对 |
| 去重 | 无需（保证唯一解）| 需要跳过重复 left / right |

````python
# ===== 通用模板：排序 + 对撞双指针（三数之和）=====
# 触发：找三元素满足和条件 + 不含重复三元组 + 允许排序
# 复杂度：O(n²) time | O(1) space（不计输出）

def three_sum_template(nums: list, target: int = 0) -> list:
    nums.sort()
    n, res = len(nums), []

    for i in range(n - 2):
        # 剪枝（target=0 时）
        if target == 0 and nums[i] > 0:
            break

        # 外层去重：向前比较（i-1），不向后！
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left, right = i + 1, n - 1
        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if total == target:
                res.append([nums[i], nums[left], nums[right]])
                # 内层去重：跳过重复，再各移一步
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1

            elif total < target:
                left += 1
            else:
                right -= 1

    return res

print("✅ 三数之和模板已加载")
````

````python
# ===== LC 15 完整解法 =====
# O(n²) time | O(1) space（不计输出）
# 排序 + 对撞双指针 + 三层去重

class Solution_15:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums = sorted(nums)   # 排序：使相同元素相邻 + 启用双指针单调性
        res = []
        n = len(nums)

        for i in range(n):
            # 剪枝：已排序，nums[i] > 0 → 后续三数之和必然 > 0
            if nums[i] > 0:
                break

            # 去重①（外层）：当前 nums[i] 与上一个相同，跳过
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left, right = i + 1, n - 1

            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    res.append([nums[i], nums[left], nums[right]])
                    # 去重②（left）：跳过 left 侧重复值
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    # 去重③（right）：跳过 right 侧重复值
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                    # 两侧各移一步，进入下一段
                    left += 1
                    right -= 1

                elif total < 0:    # 和偏小 → 右移左指针
                    left += 1
                else:              # 和偏大 → 左移右指针
                    right -= 1

        return res

print("✅ LC 15 解法已加载")
````

````python
# ===== 测试用例 =====

sol15 = Solution_15()

def normalize(triples):
    return sorted([sorted(t) for t in triples])

tests15 = [
    ([-1, 0, 1, 2, -1, -4], [[-1, -1, 2], [-1, 0, 1]]),   # 示例 1
    ([0, 1, 1],              []),                            # 示例 2：无解
    ([0, 0, 0],              [[0, 0, 0]]),                   # 示例 3：全零
    ([-2, 0, 0, 2, 2],       [[-2, 0, 2]]),                 # 重复元素去重
    ([1, 2, -2, -1],         []),                            # 无解
    ([-4, -1, -1, 0, 1, 2],  [[-1, -1, 2], [-1, 0, 1]]),   # 多重复
]

all_pass = True
for nums, expected in tests15:
    result = sol15.threeSum(nums.copy())
    ok = normalize(result) == normalize(expected)
    if not ok:
        all_pass = False
    flag = "✅" if ok else "❌"
    print(f"{flag}  nums={nums}")
    print(f"      得到={normalize(result)}  期望={normalize(expected)}")
    print()

print("🎉 全部通过！" if all_pass else "⚠️ 有用例未通过")
````

## 4. 模式对照总结

### 三题横向对比

| | LC 26 | LC 141（对照）| LC 15 |
|--|--|--|--|
| **核心题型** | 读写双指针 | 追及双指针 | 排序 + 对撞双指针 |
| **时间** | O(n) | O(n) | O(n²) |
| **空间** | O(1) | O(1) | O(1) |
| **驱动逻辑** | fast 恒前进，slow 条件前进 | fast 2步，slow 1步，利用速度差 | 外层固定降维为两数之和 |
| **核心决策** | nums[fast] ≠ nums[slow] → slow++ | slow is fast → 有环 | 三层去重 |
| **易错点** | return slow+1，不是 slow | 用 `is` 不用 `==` | 外层去重向前比较 (i-1)，不向后 |

### 📐 快慢指针两种范式对照

```
读写双指针（数组原地筛选）                 追及双指针（链表判环）
──────────────────────────────            ─────────────────────────────────
slow = 0                                  slow, fast = head, head
for fast in range(1, n):                  while fast and fast.next:
    if keep(nums[fast], nums[slow]):           slow = slow.next
        slow += 1                              fast = fast.next.next
        nums[slow] = nums[fast]                if slow is fast: return True  # 有环
return slow + 1                           return False                       # 无环

keep_condition 变体：                      Floyd 原理：
  LC 26:  nums[fast] != nums[slow]          有环时 fast 每次缩短与 slow 的距离 1 步，
  LC 80:  nums[fast] != nums[slow-1]        必在环长度以内的迭代中相遇。
  LC 283: nums[fast] != 0
```
