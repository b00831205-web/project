````python
from collections import defaultdict

def _check(name, got, expected, normalizer=None):
    try:
        a = normalizer(got) if normalizer else got
        b = normalizer(expected) if normalizer else expected
        ok = (a == b)
    except Exception as e:
        ok = False; got = f"<报错: {e}>"
    mark = "✓" if ok else "✗"
    line = f"  [{mark}] {name:30s} {'PASS' if ok else 'FAIL'}"
    if not ok:
        line += f"   得到: {got!r}   期望: {expected!r}"
    print(line)
    return ok

def _norm_groups(result):
    return sorted([sorted(g) for g in result])

print("测试工具已就绪 ✓")
````

```python
def contains_duplicate(nums):
    seen = set()
    for n in nums:
        if n in seen:
            return True
        seen.add(n)
    return False

def is_anagram(s, t):
    return len(s) == len(t) and sorted(s) == sorted(t)
    # 进阶 O(n): from collections import Counter; return Counter(s) == Counter(t)

def two_sum(nums, target):
    seen = {}                       # 值 -> 下标
    for i, n in enumerate(nums):
        if target - n in seen:
            return [seen[target - n], i]
        seen[n] = i
    return []

def group_anagrams(strs):
    g = defaultdict(list)
    for s in strs:
        g[tuple(sorted(s))].append(s)   # 排序后字符元组当 key
    return list(g.values())
```
